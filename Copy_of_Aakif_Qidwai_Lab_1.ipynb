```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

m = -0.5;
b = 0;
a = 0.0001;

df = pd.read_csv('https://raw.githubusercontent.com/tofighi/MachineLearning/master/datasets/student_marks.csv')
```

**Code in Cell below calculates the mean of the sample data:**

```python
df.iloc[:,0:2].apply(np.mean)
```

**Code in Cell below calculates the standard deviation of the sample data:**

```python
df.iloc[:,0:2].apply(np.std)
```

**Code below shows all student marks from data set and places a regression line on the figure**



```python
plt.scatter(x='Midterm mark' , y ='Final mark', data=df)
plt.xlabel('Midterm Marks')
plt.ylabel('Final Marks')
plt.title("Student marks dataset")
x = np.linspace(0, 80, 80)
x = np.floor(x)
y = m*x + b
plt.plot(x, y)
plt.show()
```

**Code below shows the implementation of the gradient and error functions for 100 iterations**

---



```python
from pandas.core.indexes.datetimes import bdate_range
Midterm_mark = df['Midterm mark']
Final_mark = df['Final mark']

def gradient_descent(iterations, x, y, m, b, a):
  n = len(x)
  for i in range(iterations-1):
    y_predicted = np.dot(x, m) + b
    md = (1/n) * sum(x * (y_predicted - y))
    bd = (1/n) * sum((y_predicted - y))
    m = m - a*md
    b = b - a*bd
  return m, b

def gradient_descent_error(iterations, x, y, m, b, a):
  cost = np.zeros(iterations)
  n = len(x)
  for i in range(iterations):
    y_predicted = np.dot(x, m) + b
    md = (1/n) * sum(x * (y_predicted - y))
    bd = (1/n) * sum((y_predicted - y))
    m = m - a*md
    b = b - a*bd
    cost[i] = (1/2*n) * sum((y_predicted - y) ** 2)

  return cost

update_m, update_b = gradient_descent(100, Midterm_mark, Final_mark, m, b, a)
update_y = update_m*x + update_b
plt.scatter(Midterm_mark, Final_mark)
plt.xlabel('Midterm Marks')
plt.ylabel('Final Marks')
plt.title("Gradient descent plot after 100 iterations")
plt.plot(x, update_y)
plt.show()

costArray = gradient_descent_error(100, Midterm_mark, Final_mark, m, b, a)


plt.plot(np.linspace(0, len(costArray), len(costArray)), costArray)
plt.xlabel('Iterations');
plt.ylabel('Error');
plt.title("Error function plot after 100 iterations")


print("After 100 iterations, updated m value is", update_m)
print("After 100 iterations, updated b value is", update_b)
```

**Code below shows the implementation of the gradient and error functions for 2000 iterations**


```python
update_m, update_b = gradient_descent(2000, Midterm_mark, Final_mark, m, b, a)
update_y = update_m* x + update_b
plt.scatter(Midterm_mark, Final_mark)
plt.xlabel('Midterm Marks')
plt.ylabel('Final Marks')
plt.title("Gradient descent plot after 2000 iterations")
plt.plot(x, update_y);

costArray2 = gradient_descent_error(2000, Midterm_mark, Final_mark, m, b, a)

fig5 = plt.figure("Figure 1")
plt.plot(np.linspace(0, len(costArray2), len(costArray2)), costArray2)
plt.xlabel('Iterations');
plt.ylabel('Error');
plt.title("Error function plot after 2000 iterations")

print("After 2000 iterations, updated m value is", update_m)
print("After 2000 iterations, updated b value is", update_b)
```
